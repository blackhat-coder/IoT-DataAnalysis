```python
outliers = {}

def treat_outliers(feature):
    lower_q = np.quantile(data[feature], 0.25)
    iqr = np.quantile(data[feature], 0.5)
    upper_q = np.quantile(data[feature], 0.75)

    out = []
    for i in data[feature]:
        if i > ( upper_q + (1.5 * iqr)):
            out.append(i)
        if i < ( lower_q - (1.5 * iqr )):
            out.append(i)

    outliers[feature] = out

treat_outliers('temperature')
treat_outliers('visibility')
treat_outliers('windSpeed')
treat_outliers('precipIntensity')
treat_outliers('dewPoint')
treat_outliers('precipProbability')
treat_outliers('apparentTemperature')
treat_outliers('pressure')
```

```python

print('LEN OF OUTLIERS IN EACH FEATURE OF WEATHER')

for i in outliers.keys():

    print(f'FEATURE: {i}   ------- > {len(outliers[i])}')
```
